# Modelling

1. Menentukan Independent (Fitur) & Dependent (Target)

    - Simple linear hanya memiliki 1 fitur

```python
X = df['YearsExperience']
y = df['Salary']
```

Standarisasi

```python
X = (X - X.mean()) / X.std()
y = (y - y.mean()) / y.std()
```

2. Split Data

- Random Split

```python
def split_data(X, y, train_size=0.8, random_state=42):
    np.random.seed(random_state)
    idx = np.random.permutation(len(X))
    split_ratio = int(train_size * len(X))

    X_train = X.iloc[idx[:split_ratio]]
    X_test  = X.iloc[idx[split_ratio:]]
    y_train = y.iloc[idx[:split_ratio]]
    y_test  = y.iloc[idx[split_ratio:]]

    return X_train, X_test, y_train, y_test


X_train, X_test, y_train, y_test = split_data(X, y)
```

## Simple Linear Regression (OLS dengan rata-rata)

- digunakan untuk 1 fitur / X

4. Slope / Koefisien
    - Koefisien adalah nilai yang menunjukkan seberapa besar pengaruh variabel X terhadap Y.

```python
slope = np.sum((X - np.mean(X)) * (y - np.mean(y))) / np.sum((X - np.mean(X))**2)
```

5. Intercept / Bias
    - nilai Y ketika X = 0

```python
intercept = np.mean(y) - slope * np.mean(X)
```

6. Mencari prediksi

```python
y_pred = slope * X_test + intercept
```

# Evaluasi

1. R-squared = menjelaskan seberapa baik model menjelaskan varisasi data
    - nilai 0 : model tidak dapat menjelaskan data sama sekali
    - nilai 1 : model mampu menjelaskan seluruh variasi data

```python
r2 = 1 - (np.sum((y_test - y_pred) ** 2) / np.sum((y_test - np.mean(y_test)) ** 2))
print("R-squared:", r2)

r2_train = 1 - (np.sum((y_train - y_pred) ** 2) / np.sum((y_train - np.mean(y_train)) ** 2))
print("R-squared (Train):", r2_train)
```

2. MSE
    - semakin kecil nilai, semakin baik kinerja model

```python
mse = np.mean((y_test - y_pred) ** 2)
print("Mean Squared Error:", mse)
```

3. MAE
    - mengukur rata-rata kesalahan absolut antara nilai prediksi dan nilai aktual

```python
mae = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error:", mae)
```
